# Average number of comparisons

```python
def avg_cmp(bst):
    n = len(bst)
    # calculate num of comps for each node
    compsByNode = [math.ceil(math.log(i+2,2)) for i in range(n)]
    # return average
    return(sum(compsByNode)/n)
```

# Maximum Height of BST

```python
def max_height(bst):
    return(len(bst))
```

# Average Height of BST

```python
def avg_height(bst):
    return(avg_cmp(bst) - 1)
```

# Max vs Avg Height of BST

```python
import random

def rand_list(n):
    retList = []
    for _ in range(n):
        retList.append(random.randint(-100,100))
    return(retList)
```

```python
# set check range
start = 1
end = 10000
step = 100

# store heights
avgHeight = []
maxHeight = []

for i in range(start,end,step):
    lst = rand_list(i)
    avgHeight.append(avg_height(lst))
    maxHeight.append(max_height(lst))
    
```

```python
import matplotlib.pyplot as plt

x = [i for i in range(start,end,step)]
plt.plot(x, avgHeight, label="average")
plt.title("Avg height BST")
plt.xlabel("size")
plt.ylabel("height")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [i for i in range(start,end,step)]
plt.plot(x, maxHeight, label="max")
plt.title("Max height BST")
plt.xlabel("size")
plt.ylabel("height")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [i for i in range(start,end,step)]
plt.plot(x, maxHeight, label="max")
plt.plot(x, avgHeight, label="average")
plt.title("Avg vs Max height BST")
plt.xlabel("size")
plt.ylabel("height")
plt.legend()
plt.show()
```

# theoretical scaling

### Average Height
It should scale at O(lgn) as the number of nodes doubles at each increase of level. Thus, the number of levels grows at a logarithmic rate.

### Max Height
It should scale at O(n) as the max height is the number of nodes (in a perfectly imbalanced tree, the tree is basically a linked list. Thus the height becomes the length of the linked list).


### Comparison to Actual
Actual scaling behavior seems to match.
