# 2. Row Space

## (a) 

$Row(A)$ is the span of $n$ vectors of length $m$.

The dimension of $Row(A)$ is determined by the number of linearly independent vectors it is made of (number of basis vectors).

The number of linearly independent vectors in $Row(A)$ is equivalent to the number of pivots in the RREF of A.

The maximal number of pivots we can have in a matrix is the minimum number of $n$ and $m$. This is because a pivot unique to both its row and column.

If $n <= m$, then the maximal dimension of $Row(A)$ is $n$, which will be a subspace of $R^m$ since $n <= m$.

If $m < n$, then the maximal dimension of $Row(A)$ is $m$, which will be a subspace of $R^m$ since $m == m$.

Thus, the dimension of $Row(A)$ is always $<= m$.   
Thus, $Row(A)$ is a subspace of $R^m$

## (b) 

The only transformations between $Row(A)$ and $Row(U)$ are row operations.   
The only row operations are multiplying by a constant, adding a multiple of another row, and swapping rows.   
Thus any row of $U$ can be expressed as a linear combination of multiples of rows of $A$.      
In other words, the mapping between $Row(A)$ and $Row(U)$ is linear.

Thus, the RREF of $Row(A)$ and $Row(U)$ is the same.   
Thus, the basis of $Row(A)$ and $Row(U)$ is the same.  
Thus $Row(A) == Row(U)$.
