Without the telescoping arm, we have: 


$ a\begin{bmatrix}
    3 \\
    1 
\end{bmatrix} = 
\begin{bmatrix}
    19 \\
    5  
\end{bmatrix} $

which translates to $3a = 19$ and $a = 5$

$3a = 19$ simplifies to $a = 19/3$

$19/3 != 5$, so we have a contradiction. Thus without the arm we cannot reach the site.

So instead, we have to check what is the closest we can get to the site and if that value is less than the length of the telescoping arm.


The multiple of $v1$ that will get us closest to $v2$ is the multiple of the projection of $v2$ on to $v1$ (since 90 degrees is the smallest angle of difference we can get between a point on a span of a vector and a point that is not on the span on a vector).

The multiple of projection:
$\frac{<3,1> \dot <19,5>}{<19,5> \dot <19,5>}$

```python
oneC_v1 = vector([3,1])
oneC_v2 = vector([19,5])

print(oneC_v1.dot_product(oneC_v2))
print(oneC_v1.dot_product(oneC_v1))
```

## (d)

The new subspace that we can reach is the span of $<3,1> \pm arm_vector$ 

The maximal potential of the arm is when it is used in a direction that is perpendicular to the current 1d span of $<3,1>$ (since the maximum angle difference we can create with another vector is 90 degrees). 


A vector of magnitude 2 that is perpendicular to $<3,1>$ can be calculated as following:

A perpendicular vector would have a dot product of 0 would satisfy the following equation:

$<a,b> \dot <3,1> = 0$

$3a + b = 0$

Expressed in terms of $b$, this would be

$b = -3a$



A vector of magnitude 2 would satisfy the following equation:

$2 = \sqrt{a^2 + b^2}$

Expressed in terms of $b$, this would be

$b = \sqrt{4 - a^2}$

Since our vector would have to satisfy both of these conditions, we know that

$-3a = \sqrt{4 - a^2}$

Simplified, this would be

$9a^2 = 4 - a^2$

$10a^2 = 4$

$a = \sqrt{\frac{4}{10}}$

$a = \sqrt{\frac{2}{5}}$


Now knowing $a$, we can find $b$

$b = -3\sqrt{\frac{2}{5}}$

Thus $arm_vector = <\sqrt{\frac{2}{5}},-3\sqrt{\frac{2}{5}}>$


```python
arm = vector([sqrt(2/5),-3*sqrt(2/5)])
one_c7 = plot(2*v1,color="black",title="The new subspace: span of <3,1> $\pm$ arm")
one_cn7 = plot(v1*-2,color="black")
one_c8 = arrow(v1,v1+arm,color='aqua',ymin=-4,ymax=4)
one_cn8 = arrow(v1,v1-arm,color='aqua')
one_c9 = arrow(0.5*v1,0.5*v1+arm,color='aqua')
one_cn9 = arrow(0.5*v1,0.5*v1-arm,color='aqua')
one_c10 = arrow([0,0],arm,color='aqua')
one_cn10 = arrow([0,0],-arm,color='aqua')
one_c11 = arrow(-0.5*v1,-0.5*v1+arm,color='aqua')
one_cn11 = arrow(-0.5*v1,-0.5*v1-arm,color='aqua')
one_c12 = arrow(-v1,-v1+arm,color='aqua')
one_cn12 = arrow(-v1,-v1-arm,color='aqua')
one_c7+one_cn7 + one_c8 + one_cn8 + one_c9 + one_cn9 + one_c10 + one_cn10 + one_c11 + one_cn11 + one_c12 + one_cn12
```
