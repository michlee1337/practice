```python
# Import any packages you need here
# Also define any variables as needed

# YOUR CODE HERE (OPTIONAL)
import copy
def goal_state(n):
        goal_state = []
        for i in range(n):
            temp = []
            for j in range(n):
                temp.append(n*i + j) 
            goal_state.append(temp)
        return goal_state
        
#Now, define the class PuzzleNode:
class PuzzleNode:
    """
    Class PuzzleNode: Provides a structure for performing A* search for the n^2-1 puzzle
    """
    # YOUR CODE HERE
    def __init__(self,state,n,fval = 0, parent=None,depth=0):
        self.state = state
        self.goal_state = goal_state(n)
        self.n = n
        self.parent = [parent]
        self.children = []
        self.depth = depth
        self.f = fval
    
    def __lt__(self,other):
        return self.f < other.f

    def check_error(self):
        # Check for acceptance of the input problem
        
        # Check if it's an n x n board
        if (len(self.state) != self.n):
            return False
        # Check if each row contain exactly n elements
        if True:
            for i in range(len(self.state)):
                if len(self.state[i]) != self.n:
                    return False
        
        if True:
            temp = []
            for i in range(self.n):
                for j in range(self.n):
                    temp.append(self.state[i][j])
            temp.sort()
            if temp != list(range(0,(self.n)**2)):
                return False

        return True
    
    def create_children(self):
        # Make a move and return a new child configuration
        i,j = self.search_for_blank_space()
        
        i0 = i-1
        j0 = j-1
        i1 = i+1
        j1 = j+1

        if i0 >= 0:
            child1_state = copy.deepcopy(self.state)
            child1_state[i][j] = child1_state[i0][j]
            child1_state[i0][j] = 0
            child1 = PuzzleNode(child1_state,self.n,parent=self,depth=self.depth+1)
            self.children.append(child1)

        if j0 >= 0:
            child2_state = copy.deepcopy(self.state)
            child2_state[i][j] = child2_state[i][j0]
            child2_state[i][j0] = 0
            child2 = PuzzleNode(child2_state,self.n,parent=self,depth=self.depth+1)
            self.children.append(child2)

        if i1 < self.n:
            child3_state = copy.deepcopy(self.state)
            child3_state[i][j] = child3_state[i1][j]
            child3_state[i1][j] = 0
            child3 = PuzzleNode(child3_state,self.n,parent=self,depth=self.depth+1)
            self.children.append(child3)

        if j1 < self.n:
            child4_state = copy.deepcopy(self.state)
            child4_state[i][j] = child4_state[i][j1]
            child4_state[i][j1] = 0
            child4 = PuzzleNode(child4_state,self.n,parent=self,depth=self.depth+1)
            self.children.append(child4)

    def visualize(self):
        # Visualize the state
        for i in range(self.n):
            for j in range(self.n):
                if self.state[i][j] == 0:
                    print("_", end='      ')
                else:
                    if self.state[i][j] > 9:
                        end = "     "
                    else:
                        end = "      "
                    print(self.state[i][j], end=end)
            print("\n")

    def search_for_blank_space(self):
        for i in range(self.n):
            for j in range(self.n):
                if self.state[i][j] == 0:
                    return (i,j)

    def __str__(self):
        return str(self.state)
```
